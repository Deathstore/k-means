```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from copy import deepcopy
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=1000, n_features=2,
 centers=3, random_state=0)
```

```python
X = np.random.rand(1000, 2)
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

```python
print(X)
```

Выбор центроид случайно

```python
centroids = [X[np.random.randint(100)], X[np.random.randint(100)], X[np.random.randint(100)]]
prev_centroids = deepcopy(centroids)
print(centroids)
```

```python
def get_nearest_centroid(point, centroids):
    return np.argmin(np.linalg.norm([point - centroid for centroid in centroids], axis=1))
```

```python
clusters = [[] for _ in range(len(centroids))]
for point in X:
  clusters[get_nearest_centroid(point, centroids)].append(point)
```

```python
for i in range(len(centroids)):
  centroids[i] = np.mean(clusters[i], axis = 0)
  #print(clusters[i])
```

```python
for cluster in clusters:
  #print(cluster)
  cluster = np.array(cluster, dtype=np.double)
  plt.scatter(cluster[:, 0], cluster[:, 1])
  #print(cluster)
#print(clusters)
#print(centroids)
```

```python
for cluster in clusters:
  #print(cluster)
  cluster = np.array(cluster, dtype=np.double)
  plt.scatter(cluster[:, 0], cluster[:, 1])
  #print(cluster)
#print(clusters)
#print(centroids)
```

```python
for i in range(len(centroids)):
  centroids[i] = np.mean(clusters[i], axis = 0)
  #print(clusters[i])
```

```python
while np.linalg.norm(np.array(prev_centroids) - np.array(centroids)) > 0.00001:
    clusters = [[] for _ in range(len(centroids))]
    for point in X:
        clusters[get_nearest_centroid(point, centroids)].append(point)
    prev_centroids = deepcopy(centroids)
    for i in range(len(centroids)):
        centroids[i] = np.mean(clusters[i], axis = 0)
    #print(clusters[i])
```

```python
for cluster in clusters:
    cluster = np.array(cluster, dtype=np.double)
    plt.scatter(cluster[:, 0], cluster[:, 1])
centroids = np.array(centroids)
plt.scatter(centroids[:,0], centroids[:,1], c = 'black')
```

```python
all =[]
for cluster in clusters:
  all += cluster
dist_matrix = [[ np.linalg.norm(all[j]-all[i]) for i in range(len(all))] for j in range(len(all))]

plt.matshow(dist_matrix) 
plt.colorbar() 
```
